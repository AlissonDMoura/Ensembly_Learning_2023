```python
import os
import numpy as np
import random

from sklearn.model_selection import train_test_split

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Set up the dictionary of dictionaries
model_dict = {}

# set the dataset path
dataset_path = r'C:\Users\aliss\Documents\last Semester\Industry Project\dataset\Digiface\Dataset'

```

### This notebook was used to train all models with a label mapping saving the models and labels into a dictionary of dictionaries.
### note that the file created after this code runs will be of more than 6GB.

```python
# loop through the subdirectories in the dataset path
for subdir in os.listdir(dataset_path):

    # get the full path of the subdirectory
    subdir_path = os.path.join(dataset_path, subdir)

    # check if the path is a directory
    if os.path.isdir(subdir_path):

        # set the path for saving the trained model
        model_path = r'C:\Users\aliss\Documents\last Semester\Industry Project\dataset\Digiface\Models\{}.h5'.format(subdir)

        # set up the image data generator
        datagen = ImageDataGenerator(rescale=1./255, validation_split=0.25)

        # get the training and validation data
        train_generator = datagen.flow_from_directory(
            subdir_path, target_size=(112, 112), batch_size=8, class_mode='categorical', subset='training')
        validation_generator = datagen.flow_from_directory(
            subdir_path, target_size=(112, 112), batch_size=8, class_mode='categorical', subset='validation')

        # set up the model architecture
        model = Sequential()
        model.add(Conv2D(32, (7, 7), activation='relu', input_shape=(112, 112, 3)))
        model.add(MaxPooling2D((2, 2)))
        model.add(Conv2D(32, (3, 3), activation='relu'))
        model.add(MaxPooling2D((2, 2)))
        model.add(Flatten())
        model.add(Dense(256, activation='relu'))
        model.add(Dense(20, activation='softmax'))
        model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

        # train the model
        model.fit(train_generator, epochs=10, validation_data=validation_generator)

        # get the label index dictionary
        label_dict = train_generator.class_indices

        # reverse the dictionary to get the label name dictionary
        label_dict = {v: k for k, v in label_dict.items()}

        # update the model dictionary
        model_dict[subdir] = {"model": model, "label_dict": label_dict}

        # save the trained model
        model.save(model_path)
```

```python
# save the model_labels dictionary to a file
labels_path = r'C:\Users\aliss\Documents\last Semester\Industry Project\dataset\Digiface\Models\labels.npy'
np.save(labels_path, model_dict)
```
